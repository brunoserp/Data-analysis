### 2. Análise Geral dos Dados de Olimpíadas de Verão - Países com mais participações, maiores campeões olímpicos, etc

#### 2.1 Quais países tiveram mais participações em Olimpíadas? 
##### Exemplo, um atleta que disputou barras e solo na ginástica numa mesma edição, está 2x na tabela e conta duas vezes nessa seleção

```python
query = """ SELECT TEAM, COUNT(TEAM) as [Qtd. Participações]
            FROM OLYMPICS_DATA 
            WHERE SEASON = 'Summer'
            GROUP BY TEAM
            ORDER BY COUNT(TEAM) DESC
            LIMIT 11
        """
df_query = pd.read_sql(query,con=engine)
df_query
```

#### 2.1.1 Em quais posições estão Rússia/URSS, China e Alemanha (unificada, Oriental e Ocidental)?

```python
query = """ SELECT 
                CASE WHEN TEAM LIKE '%Russia%' THEN 'Russia'
                     WHEN TEAM LIKE '%SOV%' THEN 'URSS'
                     WHEN TEAM LIKE '%China%' THEN 'China'
                     WHEN TEAM LIKE 'Germany%' THEN 'Germany'
                ELSE TEAM
                END AS TEAM,
                COUNT(TEAM) [Qtd. Participações]
            FROM OLYMPICS_DATA 
            WHERE   SEASON = 'Summer'
                    AND (TEAM LIKE '%SOV%'
                            OR TEAM LIKE '%Russia%'
                            OR TEAM LIKE '%China%'
                            OR TEAM LIKE 'Germany%')
            GROUP BY CASE WHEN TEAM LIKE '%Russia%' THEN 'Russia'
                          WHEN TEAM LIKE '%SOV%' THEN 'URSS'
                          WHEN TEAM LIKE '%CHINA%' THEN 'China'  
                          WHEN TEAM LIKE 'Germany%' THEN 'Germany'
                ELSE TEAM
                END
            ORDER BY COUNT(TEAM) DESC
        """
df_query = pd.read_sql(query,con=engine)
df_query
```

#### Rússia e URSS somadas as tornariam a 8ª nação com mais participações da história, com 1.693 participações
#### A Alemanha com todas as participações antes da unificação não trocaria de lugar
#### A China é a 26ª nação com mais participações em modalidades

#### 2.2 Quais países tiveram mais atletas distintos nas Olimpíadas?

#### Não há grandes mudanças quando quantificamos por atletas distintos em relação ao total de participações por categorias

#### 2.3 Quais atletas foram campeões olímpicos mais vezes?

#### Agora, sobre aproveitamento de medalhas sobre participações (seja por Olimpíada participada, ou por modalidade competida)

#### 2.4.1 Quais atletas tiveram melhor aproveitamento de medalhas de ouro comparadas às vezes que competiram?

```python
query = """ SELECT NAME, TEAM, SEX, SPORT, EVENT,
                MIN(YEAR) AS [Primeira Olimpíada],
                MAX(YEAR) AS [Última Olimpíada],
                COUNT(DISTINCT YEAR) AS [Edições Participadas],
                COUNT(MEDAL) AS [Qtd. ouros],
                COUNT(EVENT) AS [Vezes que competiu],
                COUNT(MEDAL)/COUNT(EVENT) AS [Ouros/Disputas]                
            FROM OLYMPICS_DATA 
            WHERE MEDAL='Gold'
                    AND SEASON='Summer'
            GROUP BY NAME, TEAM, SEX, SPORT, EVENT
            ORDER BY COUNT(MEDAL)/COUNT(EVENT) DESC
            LIMIT 7
        """
df_query = pd.read_sql(query,con=engine)
df_query
```

#### 2.4.2 Há vários atletas que participaram só uma vez. Vamos calcular a mesma coisa, porém para atletas que mais participaram em modalidades em Olimpíadas

```python
query = """ SELECT NAME, TEAM, SEX, SPORT,
                MIN(YEAR) AS [Primeira Olimpíada],
                MAX(YEAR) AS [Última Olimpíada],
                COUNT(DISTINCT YEAR) AS [Edições Participadas],
                COUNT(MEDAL) AS [Qtd. ouros],
                COUNT(EVENT) AS [Vezes que competiu],
                COUNT(MEDAL)/COUNT(EVENT) AS [Ouros/Disputas]                
            FROM OLYMPICS_DATA 
            WHERE MEDAL='Gold'
                    AND SEASON='Summer'
            GROUP BY NAME, TEAM, SEX, SPORT
            ORDER BY COUNT(EVENT) DESC, COUNT(MEDAL)/COUNT(EVENT) DESC
            LIMIT 7
        """
df_query = pd.read_sql(query,con=engine)
df_query
```

#### 2.4.2 Quais atletas tiveram maior média de ouros por edições olímpicas que participaram?

```python
query = """ SELECT NAME, TEAM, SEX, SPORT, 
                MIN(YEAR) AS [Primeira Olimpíada],
                MAX(YEAR) AS [Última Olimpíada],
                COUNT(DISTINCT YEAR) AS [Participações],
                COUNT(MEDAL) AS [Qtd. ouros],
                COUNT(MEDAL)/COUNT(DISTINCT YEAR) AS [Ouros/Edições Participadas]                
            FROM OLYMPICS_DATA 
            WHERE MEDAL='Gold'
                    AND SEASON='Summer'
            GROUP BY NAME, TEAM, SEX, SPORT
            ORDER BY COUNT(MEDAL)/COUNT(DISTINCT YEAR) DESC
            LIMIT 7
        """
df_query = pd.read_sql(query,con=engine)
df_query
```

#### 2.5 Quais os maiores medalhistas olímpicos?

```python
query = """ SELECT NAME, TEAM, SEX, SPORT, 
                MIN(YEAR) AS [Primeira Olimpíada],
                MAX(YEAR) AS [Última Olimpíada],
                COUNT(DISTINCT YEAR) AS [Participações],
                SUM(CASE WHEN MEDAL = 'Gold' THEN 1 ELSE 0 END) AS [Qtd. Ouro],
                SUM(CASE WHEN MEDAL = 'Silver' THEN 1 ELSE 0 END) AS [Qtd. Prata],
                SUM(CASE WHEN MEDAL = 'Bronze' THEN 1 ELSE 0 END) AS [Qtd. Bronze],
                COUNT(MEDAL) AS [Qtd. Medalhas]
            FROM OLYMPICS_DATA 
            WHERE SEASON='Summer'
            GROUP BY NAME, TEAM, SEX, SPORT
            ORDER BY COUNT(MEDAL) DESC
            LIMIT 5
        """
df_query = pd.read_sql(query,con=engine)
df_query
```

#### 2.6 Quais esportes tiveram maiores médias de altura?

```python
query = """ SELECT SPORT, ROUND(AVG(HEIGHT),1) AS [Média Altura]
            FROM OLYMPICS_DATA 
            WHERE SEASON='Summer'
            GROUP BY SPORT
            ORDER BY AVG(HEIGHT) DESC
            LIMIT 7
        """
df_query = pd.read_sql(query,con=engine)
df_query
```

#### 2.7 Esportes com maiores aumentos de média de altura a partir de 2004 comparada com as edições até 1970

```python
query = """ SELECT SPORT,
                AVG(CASE WHEN YEAR <= 1970 THEN height END) AS [Média Altura até 1970],
                AVG(CASE WHEN YEAR >= 2004 THEN height END) AS [Média Altura pós 2004],
                ROUND(((AVG(CASE WHEN YEAR >= 2004 THEN height END) - AVG(CASE WHEN YEAR <= 1970 THEN height END)) / AVG(CASE WHEN YEAR <= 1970 THEN height END)) * 100, 2) AS [Aumento da Média da Altura]
            FROM OLYMPICS_DATA 
            WHERE SEASON='Summer'
            GROUP BY SPORT
            ORDER BY ROUND(((AVG(CASE WHEN YEAR >= 2004 THEN height END) - AVG(CASE WHEN YEAR <= 1970 THEN height END)) / AVG(CASE WHEN YEAR <= 1970 THEN height END)) * 100, 2) DESC
            LIMIT 5
        """
df_query = pd.read_sql(query,con=engine)
df_query
```

### 3. Participações do Brasil

##### O país teve 1.529 participações a nível de modalidade (11º geral), e 804 atletas distintos (9º geral)

#### 3.1 Quais atletas tiveram mais participações, levando em conta as modalidades?
##### Por exemplo, se um atleta disputou uma modalidade em duas categorias (50m nado livre e 100m costas), será contado 2x

```python
query = """ SELECT NAME, TEAM, SEX, SPORT, 
                MIN(YEAR) AS [Primeira Olimpíada],
                MAX(YEAR) AS [Última Olimpíada],
                COUNT(DISTINCT EVENT) AS [Qtd. Modalidades]
            FROM OLYMPICS_DATA 
            WHERE TEAM='Brazil'
                    AND SEASON='Summer'
            GROUP BY NAME, TEAM, SEX, SPORT
            ORDER BY COUNT(DISTINCT EVENT) DESC
            LIMIT 7
        """
df_query = pd.read_sql(query,con=engine)
df_query
```

#### 3.2 Quais atletas participaram de mais edições olímpicas?
##### Por exemplo, se um atleta disputou uma modalidade em duas categorias (50m nado livre e 100m costas) numa mesma edição, será contado 1x

```python
query = """ SELECT NAME, TEAM, SEX, SPORT, 
                MIN(YEAR) AS [Primeira Olimpíada],
                MAX(YEAR) AS [Última Olimpíada],
                COUNT(DISTINCT YEAR) AS [Qtd. Edições Olímpicas]
            FROM OLYMPICS_DATA 
            WHERE TEAM='Brazil'
                    AND SEASON='Summer'
            GROUP BY NAME, TEAM, SEX, SPORT
            ORDER BY COUNT(DISTINCT YEAR) DESC
            LIMIT 7
        """
df_query = pd.read_sql(query,con=engine)
df_query
```

#### 3.3 Quais atletas mais ganharam medalhas?

```python
query = """ SELECT NAME, TEAM, SEX, SPORT, 
                MIN(YEAR) AS [Primeira Olimpíada],
                MAX(YEAR) AS [Última Olimpíada],
                COUNT(DISTINCT EVENT) AS [Qtd. Modalidades],
                SUM(CASE WHEN MEDAL = 'Gold' THEN 1 ELSE 0 END) AS [Qtd. Ouro],
                SUM(CASE WHEN MEDAL = 'Silver' THEN 1 ELSE 0 END) AS [Qtd. Prata],
                SUM(CASE WHEN MEDAL = 'Bronze' THEN 1 ELSE 0 END) AS [Qtd. Bronze],
                COUNT(MEDAL) AS [Qtd. Medalhas]
            FROM OLYMPICS_DATA 
            WHERE TEAM='Brazil'
                    AND SEASON='Summer'
            GROUP BY NAME, TEAM, SEX, SPORT
            ORDER BY COUNT(MEDAL) DESC
            LIMIT 7
        """
df_query = pd.read_sql(query,con=engine)
df_query
```

#### 3.4 Quais atletas foram mais vezes campeões olímpicos?

```python
query = """ SELECT NAME, TEAM, SEX, SPORT, 
                MIN(YEAR) AS [Primeira Olimpíada],
                MAX(YEAR) AS [Última Olimpíada],
                COUNT(DISTINCT EVENT) AS [Qtd. Modalidades],
                COUNT(MEDAL) AS [Qtd. Ouros]
            FROM OLYMPICS_DATA 
            WHERE TEAM='Brazil' 
                    AND MEDAL='Gold'
                    AND SEASON='Summer'
            GROUP BY NAME, TEAM, SEX, SPORT
            ORDER BY COUNT(MEDAL) DESC
            LIMIT 7
        """
df_query = pd.read_sql(query,con=engine)
df_query
```

#### 3.5 Quantos atletas ganharam ao menos uma medalha?

```python
query = """ SELECT COUNT(NAME) AS [Qtd Medalhistas]
            FROM OLYMPICS_DATA 
            WHERE TEAM='Brazil' 
                AND MEDAL != 'None'
                AND SEASON = 'Summer'
        """
df_query = pd.read_sql(query,con=engine)
df_query
```
